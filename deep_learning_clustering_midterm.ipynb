# Task
Implement a Deep Clustering Network (DCN) model. This involves defining the `AutoEncoder` and `DCN` classes, implementing reconstruction and clustering loss functions, initializing cluster centroids using KMeans on `X_scaled`, setting up and running the training loop for 100 epochs with Adam optimizer (lr=1e-3) and a custom lambda for clustering loss, and finally evaluating the clustering using the silhouette score. Additionally, visualize the clusters in 2D using PCA and save the trained `AutoEncoder` model and centroids.

## Define AutoEncoder and DCN Models

### Subtask:
Implement the `AutoEncoder` class with the specified architecture (128 -> 64 -> 10 latent dimension, symmetric decoder, ReLU activations) and the `DCN` class which incorporates the autoencoder and manages cluster centroids.


**Reasoning**:
The subtask requires defining the `AutoEncoder` and `DCN` classes based on the provided architecture and specifications. This step imports the necessary PyTorch libraries and implements both classes, including their `__init__` and `forward` methods as detailed in the instructions.



```python
import torch
import torch.nn as nn

class AutoEncoder(nn.Module):
    def __init__(self, input_dim=17, latent_dim=10):
        super(AutoEncoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, latent_dim)
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, input_dim)
        )

    def forward(self, x):
        latent = self.encoder(x)
        reconstructed = self.decoder(latent)
        return reconstructed, latent

class DCN(nn.Module):
    def __init__(self, autoencoder, num_clusters, latent_dim=10):
        super(DCN, self).__init__()
        self.autoencoder = autoencoder
        self.num_clusters = num_clusters
        # Initialize cluster centroids randomly
        self.centroids = nn.Parameter(torch.rand(num_clusters, latent_dim))

    def forward(self, x):
        reconstructed, latent_embedding = self.autoencoder(x)
        return reconstructed, latent_embedding

print("AutoEncoder and DCN classes defined successfully.")
```

## Implement DCN Loss Functions

### Subtask:
Define the reconstruction loss using `nn.MSELoss` and a custom clustering loss function based on the provided formula: `||Z - mu_c||^2`, where Z is the latent embedding and mu_c is the cluster centroid.


**Reasoning**:
The subtask requires defining the reconstruction loss and a custom clustering loss function. I will create a code block to define `nn.MSELoss` for reconstruction and then implement the custom clustering loss function as described, calculating the squared Euclidean distance to the closest centroid.



```python
reconstruction_loss_fn = nn.MSELoss()

def clustering_loss_fn(latent_embedding, centroids):
    # Calculate squared Euclidean distance between each latent_embedding and all centroids
    # latent_embedding: (batch_size, latent_dim)
    # centroids: (num_clusters, latent_dim)

    # Expand dimensions for broadcasting: (batch_size, 1, latent_dim) - (1, num_clusters, latent_dim)
    # This results in (batch_size, num_clusters, latent_dim)
    diff = latent_embedding.unsqueeze(1) - centroids.unsqueeze(0)

    # Square the differences and sum along the latent_dim to get squared Euclidean distances
    # Resulting shape: (batch_size, num_clusters)
    squared_distances = torch.sum(diff ** 2, dim=2)

    # For each latent_embedding, find the minimum squared distance to any centroid
    # Resulting shape: (batch_size,)
    min_squared_distances, _ = torch.min(squared_distances, dim=1)

    return torch.mean(min_squared_distances)

print("Reconstruction and clustering loss functions defined.")
```

## Prepare Data and Initialize Centroids

### Subtask:
Convert `X_scaled` (the preprocessed dataset) into a PyTorch Tensor and create a DataLoader. Initialize the cluster centroids by performing KMeans on the latent embeddings obtained from a pre-trained or initially randomly initialized autoencoder.


**Reasoning**:
The subtask requires converting `X_scaled` to a PyTorch tensor, creating a DataLoader, instantiating the `AutoEncoder` and `DCN` models, setting the device, generating latent embeddings, performing KMeans on these embeddings, and finally updating the `dcn.centroids` with the KMeans results. All these steps are interconnected and should be executed sequentially.



```python
import numpy as np
from sklearn.cluster import KMeans
from torch.utils.data import TensorDataset, DataLoader

# 1. Convert X_scaled to PyTorch Tensor
X_scaled_tensor = torch.from_numpy(X_scaled).float()

# Get input_dim from X_scaled
input_dim = X_scaled_tensor.shape[1]
num_clusters = 10 # As specified in the instructions

# 2. Create TensorDataset and DataLoader
dataset = TensorDataset(X_scaled_tensor)
dataloader = DataLoader(dataset, batch_size=256, shuffle=True)

# 3. Instantiate AutoEncoder and DCN models
autoencoder = AutoEncoder(input_dim=input_dim, latent_dim=10)
dcn = DCN(autoencoder, num_clusters=num_clusters, latent_dim=10)

# 4. Set device and move models
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
autoencoder.to(device)
dcn.to(device)

print(f"Using device: {device}")

# 5. Pass all data through the autoencoder.encoder to get the latent embeddings
# It's better to get all latent embeddings at once for KMeans initialization
# Put model in eval mode to prevent dropout/batchnorm from affecting initial embeddings
autoencoder.eval()
latent_embeddings_list = []
with torch.no_grad(): # No need to calculate gradients for initialization
    for batch in dataloader:
        data = batch[0].to(device)
        _, latent = autoencoder(data)
        latent_embeddings_list.append(latent.cpu().numpy())

initial_latent_embeddings = np.vstack(latent_embeddings_list)

# 6. Import KMeans and fit a KMeans model
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
kmeans.fit(initial_latent_embeddings)

# 7. Update dcn.centroids with the cluster centers from KMeans
dcn.centroids.data = torch.from_numpy(kmeans.cluster_centers_).float().to(device)

print("Data prepared, DataLoader created, models instantiated and moved to device.")
print("Initial cluster centroids set using KMeans on initial latent embeddings.")
```

## Set up Training Loop

### Subtask:
Implement the end-to-end training loop for the DCN model. This includes setting up the optimizer (Adam, lr=1e-3), device auto-detection (CPU/GPU), iterating for 100 epochs, calculating total loss (reconstruction + lambda * clustering loss), backpropagation, and weight updates. Centroids will be updated per epoch after processing all batches.


**Reasoning**:
The subtask requires setting up and running the training loop for the DCN model, which includes defining the optimizer, calculating losses, performing backpropagation, updating weights, and re-initializing centroids using KMeans after each epoch. This code block will implement all these steps as instructed.



```python
import torch.optim as optim

# 1. Set the DCN model to training mode
dcn.train()
dcn.autoencoder.train() # Ensure autoencoder part is also in train mode

# 2. Define the Adam optimizer
optimizer = optim.Adam(dcn.parameters(), lr=1e-3)

# 3. Define lambda_val for clustering loss contribution
lambda_val = 0.1

# 4. Create an empty list to store the total loss for each epoch
epoch_losses = []

# 5. Loop for 100 epochs
num_epochs = 100
print(f"Starting DCN training for {num_epochs} epochs...")

for epoch in range(num_epochs):
    current_epoch_losses = []
    for batch_idx, batch in enumerate(dataloader):
        data = batch[0].to(device)

        # c. Zero out the gradients of the optimizer
        optimizer.zero_grad()

        # d. Pass the input data through the DCN model
        reconstructed, latent_embedding = dcn(data)

        # e. Calculate the reconstruction loss
        recon_loss = reconstruction_loss_fn(reconstructed, data)

        # f. Calculate the clustering loss
        clust_loss = clustering_loss_fn(latent_embedding, dcn.centroids)

        # g. Calculate the total loss
        total_loss = recon_loss + lambda_val * clust_loss

        # h. Perform backpropagation
        total_loss.backward()

        # i. Update the model's weights
        optimizer.step()
        current_epoch_losses.append(total_loss.item())

    # j. After processing all batches in an epoch, update the DCN's centroids
    dcn.eval() # Set DCN to eval mode to get consistent latent embeddings for KMeans
    dcn.autoencoder.eval()

    all_latent_embeddings = []
    with torch.no_grad():
        for batch in dataloader:
            data = batch[0].to(device)
            _, latent = dcn(data)
            all_latent_embeddings.append(latent.cpu().numpy())

    current_latent_embeddings = np.vstack(all_latent_embeddings)

    # Perform KMeans on these embeddings
    kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
    kmeans.fit(current_latent_embeddings)

    # Update dcn.centroids with the new cluster centers
    dcn.centroids.data = torch.from_numpy(kmeans.cluster_centers_).float().to(device)

    dcn.train() # Set DCN back to train mode
    dcn.autoencoder.train()

    # k. Append the epoch's average total loss
    avg_epoch_loss = sum(current_epoch_losses) / len(current_epoch_losses)
    epoch_losses.append(avg_epoch_loss)

    # 6. Print the average total loss for each epoch
    print(f"Epoch {epoch+1}/{num_epochs}, Average Total Loss: {avg_epoch_loss:.4f}")

print("Training complete.")
```

**Reasoning**:
The training loop has completed. Now, to evaluate the clustering performance, I need to obtain the final latent embeddings from the trained DCN model, assign cluster labels based on the final centroids, and then calculate the silhouette score using these labels and embeddings.



```python
from sklearn.metrics import silhouette_score

# Set the DCN model to evaluation mode
dcn.eval()
dcn.autoencoder.eval()

# Get the final latent embeddings and assign cluster predictions
final_latent_embeddings = []
with torch.no_grad():
    for batch in dataloader:
        data = batch[0].to(device)
        _, latent = dcn(data)
        final_latent_embeddings.append(latent.cpu().numpy())

final_latent_embeddings = np.vstack(final_latent_embeddings)

# Get the final centroids from the DCN model
final_centroids = dcn.centroids.cpu().detach().numpy()

# Predict cluster assignments for each latent embedding based on the closest centroid
def assign_clusters(latent_embeddings, centroids):
    distances = np.sqrt(((latent_embeddings[:, np.newaxis, :] - centroids[np.newaxis, :, :]) ** 2).sum(axis=2))
    return np.argmin(distances, axis=1)

cluster_assignments = assign_clusters(final_latent_embeddings, final_centroids)

# Calculate the Silhouette Score
silhouette_avg = silhouette_score(final_latent_embeddings, cluster_assignments)

print(f"Final Silhouette Score: {silhouette_avg:.4f}")
```

**Reasoning**:
To fulfill the remaining requirements of the main task, I will now visualize the clusters in 2D using PCA, save the trained AutoEncoder model and the learned centroids.



```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1. Visualize clusters in 2D using PCA

# Perform PCA to reduce latent embeddings to 2 dimensions
pca = PCA(n_components=2)
latent_2d = pca.fit_transform(final_latent_embeddings)

# Plot the 2D latent space with clusters colored
plt.figure(figsize=(10, 8))
scatter = plt.scatter(latent_2d[:, 0], latent_2d[:, 1], c=cluster_assignments, cmap='viridis', s=50, alpha=0.6)

# Plot the centroids as well
centroids_2d = pca.transform(final_centroids)
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1], c='red', marker='X', s=200, label='Centroids', edgecolor='black')

plt.title('DCN Clusters in 2D PCA Space')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.colorbar(scatter, label='Cluster Assignment')
plt.grid(True)
plt.show()

# 2. Save the trained AutoEncoder model
torch.save(dcn.autoencoder.state_dict(), 'autoencoder_model.pth')
print("AutoEncoder model saved successfully as 'autoencoder_model.pth'.")

# 3. Save the learned centroids
torch.save(dcn.centroids.data, 'dcn_centroids.pth')
print("DCN centroids saved successfully as 'dcn_centroids.pth'.")
```
